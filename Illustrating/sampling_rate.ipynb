```python
import numpy as np
import matplotlib.pyplot as plt

# To tn hiu mu
fs = 1000  # Tn s ly mu 1000 Hz
t = np.linspace(0, 1, fs, endpoint=False)
signal = np.sin(2 * np.pi * 50 * t) + np.sin(2 * np.pi * 120 * t)

# Hin th tn hiu
plt.figure(figsize=(20, 4))
plt.plot(t, signal)
plt.title("Tn hiu vi tn s ly mu 1000 Hz")
plt.xlabel("Thi gian (s)")
plt.ylabel("Bin ")
plt.show()
```
